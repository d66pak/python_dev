```python
sumrow_series = exceldf[['Jan', 'Feb', 'Mar', 'Total']].sum()
sumrow_series
```

```python
# Convert series to DF
sumrowdf = pd.DataFrame(data=sumrow_series)
sumrowdf.head()
```

```python
# Transpose the DF
sumrowdf = sumrowdf.T
sumrowdf.head()
```

```python
# Add other columns to this DF
sumrowdf = sumrowdf.reindex(columns=exceldf.columns)
sumrowdf.head()
```

```python
# Append it to final DF
excelsumdf = exceldf.append(sumrowdf, ignore_index=True)
excelsumdf.tail()
```

```python
process.extractOne('Minnesotta', choices=state_to_code.keys())
```

```python
process.extractOne('AllaBBaaamaaa', choices=state_to_code.keys())
```

```python
process.extractOne('AllaBBaaamaaa', choices=state_to_code.keys(), score_cutoff=80)
```

```python
process.extractOne('Delaware', choices=state_to_code.keys(), score_cutoff=80)
```

```python
# Transform function
def convert_state(row):
    '''
    row: Series
    '''
    abbrev = process.extractOne(str(row['state']), choices=state_to_code.keys(), score_cutoff=80)
    if abbrev:
        return state_to_code[abbrev[0]]
    return np.nan
```

```python
# Insert blank column
excelsumdf.insert(6, 'abbrev', np.nan)
excelsumdf.tail()
```

```python
excelsumdf['abbrev'] = excelsumdf.apply(convert_state, axis=1)
excelsumdf.tail()
```

### 4. Subtotals by state

```python
excelsubtotaldf = excelsumdf[['abbrev', 'Jan', 'Feb', 'Mar', 'Total']].groupby('abbrev').sum()
excelsubtotaldf.head()
```

```python
excelsubtotaldf.tail()
```

```python
# Format
def money(x):
    return "${:,.0f}".format(x)
```

```python
excelsubtotaldf = excelsubtotaldf.applymap(money)
excelsubtotaldf.head()
```
