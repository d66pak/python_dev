```python
def compare_df(dfA,dfB):
    """Compare the two indexed Dataframes based on index column values"""
    indexAminusB = (dfA.index.difference(dfB.index))
    indexBminusA = (dfB.index.difference(dfA.index))
    indexIntersect = dfA.index.intersection(dfB.index)

    dfA_minus_dfB = dfA.ix[indexAminusB]
    dfB_minus_dfA = dfB.ix[indexBminusA]
    dfA_Intersect = dfA.ix[indexIntersect]
    dfB_Intersect = dfB.ix[indexIntersect]

    dfA_minus_dfB.sort_index(inplace=True)
    dfB_minus_dfA.sort_index(inplace=True)
    dfA_Intersect.sort_index(inplace=True)
    dfB_Intersect.sort_index(inplace=True)

    return dfA_minus_dfB,dfA_Intersect,dfB_Intersect,dfB_minus_dfA
```

# Read CSV file and create summary

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataIn/Redshift-Vaccume-summary.csv')
```

```python
df.head()
```

```python
df['vac_duration_mins'] = df['vac_duration_secs']/60
```

```python
df = df[df['vac_start_status'] == 'Started Reindex']
```

```python
df.head()
```

```python
df_small = df[['tablename', 'vac_duration_mins', 'vac_end_rows']]
```

```python
df_small.head()
```

### https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/

```python
gb_summary = df_small.groupby(['tablename'])
```

```python
gb_summary.groups.keys()
```

```python
gb_summary.groups['orders_orderlineitems']
```

```python
gb_summary.first()
```

```python
df_summary = gb_summary.agg({'vac_duration_mins': 'mean', 'vac_end_rows': 'mean'})
```

```python
df_summary.head()
```

```python
df_summary.sort_values(by=['vac_duration_mins'], ascending=False)
```

```python
df_summary.sort_values(by=['vac_duration_mins'], ascending=False).round({'vac_end_rows': 0, 'vac_duration_mins': 2}).to_csv('dataOut/vacuum-summary.csv')
```
