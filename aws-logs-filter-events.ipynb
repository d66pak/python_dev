```python
import os
import boto3
import time
from datetime import datetime
from datetime import timedelta
from botocore.compat import json, total_seconds
```

```python
aws_profile = 'default'
# os.environ['AWS_PROFILE'] = aws_profile
logs = boto3.session.Session(profile_name=aws_profile).client('logs')
# logs = boto3.client('logs')
```

```python
kwargs = {
    'logGroupName': '/aws/lambda/CustomerSales',
    'interleaved': True,
    'startTime': int(total_seconds((datetime.utcnow() - timedelta(seconds=300)) - datetime(1970, 1, 1))) * 1000,
    'endTime': int(total_seconds(datetime.utcnow() - datetime(1970, 1, 1))) * 1000
}
```

```python
kwargs
```

```python
count = 0
while True:
    count += 1
    print kwargs
    d_response = logs.filter_log_events(**kwargs)
    if 'events' in d_response and d_response['events']:
        print json.dumps(d_response)
        break
    elif 'nextToken' in d_response and d_response['nextToken']:
        kwargs['nextToken'] = d_response['nextToken']
    else:
        print 'Error: no nextToken found!'
        print json.dumps(d_response)
        break
        
            
```
